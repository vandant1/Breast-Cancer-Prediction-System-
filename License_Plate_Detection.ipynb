```python
import os

# Paths
image_folder = "/content/images"
label_folder = "/content/labels"

# Get lists of files
image_files = set(f.split('.')[0] for f in os.listdir(image_folder) if f.endswith(('.png', '.jpg', '.jpeg')))
label_files = set(f.split('.')[0] for f in os.listdir(label_folder) if f.endswith('.txt'))

# Identify unmatched files
images_without_labels = image_files - label_files
labels_without_images = label_files - image_files

# Delete unmatched files
for img in images_without_labels:
    img_path = os.path.join(image_folder, f"{img}.png")
    if not os.path.exists(img_path):  # Handle jpg/jpeg cases
        img_path = os.path.join(image_folder, f"{img}.jpg")
    os.remove(img_path)
    print(f"Removed image without label: {img_path}")

for lbl in labels_without_images:
    lbl_path = os.path.join(label_folder, f"{lbl}.txt")
    os.remove(lbl_path)
    print(f"Removed label without image: {lbl_path}")

print("Dataset cleanup completed!")
```

```python
import os
import cv2
import matplotlib.pyplot as plt

def visualize_sample(image_path, label_path):
    # Load image
    image = cv2.imread(image_path)
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

    # Load label
    with open(label_path, 'r') as file:
        lines = file.readlines()

    # Draw bounding boxes
    for line in lines:
        values = line.strip().split()
        class_id, x_center, y_center, width, height = map(float, values)

        h, w, _ = image.shape
        x1 = int((x_center - width / 2) * w)
        y1 = int((y_center - height / 2) * h)
        x2 = int((x_center + width / 2) * w)
        y2 = int((y_center + height / 2) * h)

        cv2.rectangle(image, (x1, y1), (x2, y2), (255, 0, 0), 2)

    # Show the image
    plt.imshow(image)
    plt.axis('off')
    plt.show()

# Path to folders
image_folder = '/content/images'
label_folder = '/content/labels'

# Get the list of files
image_files = sorted(os.listdir(image_folder))  # List and sort image files
label_files = sorted(os.listdir(label_folder))  # List and sort label files

# Ensure there are files in both folders
if image_files and label_files:
    # Dynamically construct paths for the first image and label file
    image_path = os.path.join(image_folder, image_files[0])  # First image file
    label_path = os.path.join(label_folder, label_files[0])  # First label file

    # Call the function with the dynamic paths
    visualize_sample(image_path, label_path)
else:
    print("Either the images or labels folder is empty.")
```

```python
def validate_annotations(image_folder, label_folder):
    issues = []
    for image_file in os.listdir(image_folder):
        label_file = os.path.join(label_folder, f"{os.path.splitext(image_file)[0]}.txt")
        image_path = os.path.join(image_folder, image_file)

        # Validate if label file exists
        if not os.path.exists(label_file):
            issues.append(f"Missing label for {image_file}")
            continue

        # Check bounding box coordinates
        image = cv2.imread(image_path)
        h, w, _ = image.shape
        with open(label_file, 'r') as file:
            for line in file:
                _, x_center, y_center, width, height = map(float, line.strip().split())
                if not (0 <= x_center <= 1 and 0 <= y_center <= 1 and 0 <= width <= 1 and 0 <= height <= 1):
                    issues.append(f"Invalid bounding box in {label_file}")

    if issues:
        print("Issues found:")
        for issue in issues:
            print(issue)
    else:
        print("All annotations are valid.")

validate_annotations(image_folder, label_folder)
```

```python
import os
import shutil
from sklearn.model_selection import train_test_split

# Paths
image_folder = '/content/images'
label_folder = '/content/labels'

# Create directories for splits
for split in ['train', 'val', 'test']:
    os.makedirs(f"data/{split}/images", exist_ok=True)
    os.makedirs(f"data/{split}/labels", exist_ok=True)

# Get list of image files
image_files = [f for f in os.listdir(image_folder) if f.endswith('.png')]

# Split data
train_files, test_files = train_test_split(image_files, test_size=0.2, random_state=42)
val_files, test_files = train_test_split(test_files, test_size=0.5, random_state=42)

# Helper function to move files
def move_files(file_list, split):
    for file in file_list:
        shutil.copy(os.path.join(image_folder, file), f"data/{split}/images/{file}")
        label_file = f"{os.path.splitext(file)[0]}.txt"
        shutil.copy(os.path.join(label_folder, label_file), f"data/{split}/labels/{label_file}")

# Move files
move_files(train_files, 'train')
move_files(val_files, 'val')
move_files(test_files, 'test')

print("Dataset split completed!")

```

```python
from IPython.display import display, Javascript
from google.colab.output import eval_js
from base64 import b64decode

def take_photo(filename='photo.jpg', quality=0.8):
    js = Javascript('''
        async function takePhoto(quality) {
            const div = document.createElement('div');
            const capture = document.createElement('button');
            capture.textContent = 'Capture Photo';
            div.appendChild(capture);

            const video = document.createElement('video');
            video.style.display = 'block';
            div.appendChild(video);

            document.body.appendChild(div);

            const stream = await navigator.mediaDevices.getUserMedia({video: true});
            video.srcObject = stream;
            await video.play();

            await new Promise((resolve) => capture.onclick = resolve);

            const canvas = document.createElement('canvas');
            canvas.width = video.videoWidth;
            canvas.height = video.videoHeight;
            canvas.getContext('2d').drawImage(video, 0, 0);
            stream.getTracks().forEach(track => track.stop());
            div.remove();

            return canvas.toDataURL('image/jpeg', quality);
        }
        ''')
    display(js)
    binary = b64decode(eval_js('takePhoto({})'.format(quality)).split(',')[1])
    with open(filename, 'wb') as f:
        f.write(binary)
    return filename

filename = take_photo()
print(f"Saved photo to {filename}")
```

```python
import cv2
import torch
from yolov5.models.common import DetectMultiBackend
from yolov5.utils.general import non_max_suppression, scale_coords
from yolov5.utils.plots import Annotator
from yolov5.utils.torch_utils import select_device

# Load model
device = select_device('cuda' if torch.cuda.is_available() else 'cpu')
model = DetectMultiBackend('runs/train/exp/weights/best.pt', device=device)

# Open webcam
cap = cv2.VideoCapture(0)  # Use 0 for primary webcam
if not cap.isOpened():
    print("Error: Could not open webcam.")
    exit()

# Process webcam feed
while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        print("Error: Failed to capture frame.")
        break

    # Convert to tensor and perform detection
    img = torch.from_numpy(frame).permute(2, 0, 1).float().to(device) / 255.0
    img = img.unsqueeze(0)
    pred = model(img)

    # Apply Non-Max Suppression
    pred = non_max_suppression(pred)

    # Draw bounding boxes
    annotator = Annotator(frame, line_width=2)
    for det in pred:
        if det is not None and len(det):
            det[:, :4] = scale_coords(img.shape[2:], det[:, :4], frame.shape).round()
            for *xyxy, conf, cls in det:
                label = f"{model.names[int(cls)]} {conf:.2f}"
                annotator.box_label(xyxy, label, color=(255, 0, 0))

    # Show frame
    cv2.imshow('YOLOv5 Webcam Detection', frame)

    # Exit on 'q' key
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()


```
